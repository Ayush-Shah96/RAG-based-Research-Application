# Build FAISS Index

Run this notebook **once** (or whenever you add new documents to `data/`) to:
1. Load all documents from `data/`
2. Split them into chunks
3. Embed them with `sentence-transformers/all-MiniLM-L6-v2`
4. Save the FAISS index to `vectorstore/`

```python
import sys, os
# Ensure the project root is on the path
sys.path.insert(0, os.path.abspath('..'))
os.chdir('..')  # run from project root
```

```python
from src.processing.loader import load_documents
from src.processing.splitter import split_documents
from src.processing.embeddings import build_vectorstore

# Step 1 — Load
docs = load_documents()
print(f"Loaded {len(docs)} document(s)")
```

```python
# Step 2 — Split
chunks = split_documents(docs)
print(f"Created {len(chunks)} chunk(s)")
```

```python
# Step 3 — Embed & Save
vectorstore = build_vectorstore(chunks)
print("Vectorstore built and saved successfully!")
```

```python
# Quick sanity check
results = vectorstore.similarity_search("test query", k=2)
for r in results:
    print(r.metadata, r.page_content[:120])
```
